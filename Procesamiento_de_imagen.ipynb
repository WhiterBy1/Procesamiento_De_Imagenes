```python
from skimage import io 
import matplotlib.pyplot as plt 
# Cargar imagen 
image = io.imread('images/RM-neuro.png') 
# Separar canales 
R, G, B = image[:,:,0], image[:,:,1], image[:,:,2] 
# Visualizar canales 
fig, ax = plt.subplots(1, 3, figsize=(15, 5)) 
ax[0].imshow(R, cmap='Reds') 
ax[1].imshow(G, cmap='Greens') 
ax[2].imshow(B, cmap='Blues') 
for a in ax: 
    a.axis('off') 
plt.show() 
```

```python
import matplotlib.pyplot as plt
from skimage import io
from skimage.filters import threshold_otsu
from skimage.color import rgb2gray

# Load the image
image = io.imread('images/Imagen (156).PNG')

# Convert to grayscale if the image is RGB
if len(image.shape) == 3:
    image = rgb2gray(image)

# Compute Otsu's threshold
thresh = threshold_otsu(image)

# Create binary image
binary = image > thresh

# Visualize the result
plt.figure(figsize=(5, 5))
plt.imshow(binary, cmap='gray')
plt.axis('off')
plt.title('Otsu Thresholding')
plt.show()
```

```python
from skimage.filters import median 
from skimage.morphology import disk 
# Filtro mediano 
filtered_image = median(image, disk(3)) 
# Visualizar imagen filtrada 
plt.figure(figsize=(5, 5)) 
plt.imshow(filtered_image, cmap='gray') 
plt.axis('off') 
plt.show() 
```

```python
import cv2
from skimage import io
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu

# Leer la imagen
image = io.imread('images/Dall.jpg')

# Convertir a escala de grises si es necesario
if len(image.shape) == 3:
    image = rgb2gray(image)

# Calcular el umbral de Otsu
thresh = threshold_otsu(image)

# Crear la imagen binaria
binary = (image > thresh).astype('uint8') * 255

# Mostrar la imagen usando OpenCV
cv2.imshow('Imagen Binaria', binary)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
